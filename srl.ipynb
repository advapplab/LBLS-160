```python
### MD5
def md5(i):
    hl = hb.md5()
    
    hl.update(str(i).encode(encoding='utf-8'))

    #print('加密前為 ：' + str(i))
    #print('加密後為 ：' + hl.hexdigest())
    
    return hl.hexdigest()
```

```python
def col_1102ncu(x, t):
    
    if x != 'userid':
        
        if t == 'm':
            x = list(srl_1102ncu_motivation).index(x)
            x = 'srl_m_'+str(x)
            
        else:
            x = list(srl_1102ncu_strategy).index(x)
            x = 'srl_s_'+str(x)

    return x
```

```python
def col_1092ncu(x):
    if x != 'userid':
        x = list(srl_1092ncu_motivation_b).index(x)
        x = 'srl_m_' + str(x)
        
    return x
```

```python
def col_1101ntut(x, t):
    
    if x != 'userid':
        
        if t == 'm':
            x = list(srl_1101ntut_motivation).index(x)
            x = 'srl_m_'+str(x)
        else:
        
            x = list(srl_1101ntut_strategy).index(x)
            x = 'srl_s_'+str(x)

    return x
```

```python
def col(x, t):
    
    if x != 'userid':
        x = x.split('.')[0]
        x = 'srl_'+t+'_'+x

    return x
```
