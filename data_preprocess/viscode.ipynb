```python
import pandas as pd
import hashlib as hb
```

```python
viscode_1092ncu = pd.read_csv('./data/1092_NCU/1092_viscode_logs.csv')
# viscode_1101ntnu = pd.read_csv('./data/1101_NTUT/1101-北科_viscodeLog.csv')
viscode_1102ncu = pd.read_csv('./data/1102_NCU/1102_viscodeLog.csv')
```

```python
viscode_1102ncu['codeLength']
```

```python
### MD5
def md5(i):
    hl = hb.md5()
    
    hl.update(str(i).encode(encoding='utf-8'))

    #print('加密前為 ：' + str(i))
    #print('加密後為 ：' + hl.hexdigest())
    
    return hl.hexdigest()
```

```python
viscode_1092ncu['userid'] = viscode_1092ncu['username'].astype(str)
viscode_1092ncu['userid'] = viscode_1092ncu['username'].apply(md5)

# viscode_1101ntnu['userid'] = viscode_1101ntnu['username'].astype(str)
# viscode_1101ntnu['userid'] = viscode_1101ntnu['username'].apply(md5)

viscode_1102ncu['userid'] = viscode_1102ncu['username'].astype(str)
viscode_1102ncu['userid'] = viscode_1102ncu['username'].apply(md5)
```

```python
viscode_1092ncu = viscode_1092ncu.drop('username', 1)
# viscode_1101ntnu = viscode_1101ntnu.drop('username', 1)
viscode_1102ncu = viscode_1102ncu.drop('username', 1)
```

```python
keep_list = ['userid', 'event', 'codeLength', 'errorName']

viscode_1092ncu = viscode_1092ncu[keep_list]
# viscode_1101ntnu = viscode_1101ntnu[keep_list]
viscode_1102ncu = viscode_1102ncu[keep_list]

# viscode_pd = 
```

```python
def comma_to_int (x):
    x = str(x)
    x = x.replace(',', '')
    int(x)
    
    return x
```

```python
viscode_1092ncu_event_grouped = viscode_1092ncu.groupby(['userid', 'event']).size().reset_index()
viscode_1092ncu_event_grouped = viscode_1092ncu_event_grouped.pivot_table(0, ['userid'], 'event').fillna(0).astype('int')

viscode_1092ncu_error_grouped = viscode_1092ncu.groupby(['userid', 'errorName']).size().reset_index()
viscode_1092ncu_error_grouped = viscode_1092ncu_error_grouped.pivot_table(0, ['userid'], 'errorName').fillna(0).astype('int')


viscode_1092ncu['codeLength'] = viscode_1092ncu['codeLength'].fillna(0).apply(comma_to_int)
viscode_1092ncu_codelen_grouped = viscode_1092ncu[['userid', 'codeLength']].groupby('userid').max()
```

```python
viscode_1102ncu_event_grouped = viscode_1102ncu.groupby(['userid', 'event']).size().reset_index()
viscode_1102ncu_event_grouped = viscode_1102ncu_event_grouped.pivot_table(0, ['userid'], 'event').fillna(0).astype('int')

viscode_1102ncu_error_grouped = viscode_1102ncu.groupby(['userid', 'errorName']).size().reset_index()
viscode_1102ncu_error_grouped = viscode_1102ncu_error_grouped.pivot_table(0, ['userid'], 'errorName').fillna(0).astype('int')

# viscode_1102ncu['codeLength'] = viscode_1102ncu['codeLength'].fillna(0).apply(comma_to_int)
viscode_1102ncu_codelen_grouped = viscode_1102ncu[['userid', 'codeLength']].groupby('userid').max()
```

```python
viscode_1092ncu_merged_pd = pd.concat([viscode_1092ncu_event_grouped, 
                                       viscode_1092ncu_error_grouped, 
                                       viscode_1092ncu_codelen_grouped], axis=1)

viscode_1102ncu_merged_pd = pd.concat([viscode_1102ncu_event_grouped, 
                                       viscode_1102ncu_error_grouped, 
                                       viscode_1102ncu_codelen_grouped], axis=1)
```

```python
viscode_1092ncu_merged_pd = viscode_1092ncu_merged_pd.fillna(0)
viscode_1102ncu_merged_pd = viscode_1102ncu_merged_pd.fillna(0)
```

```python
viscode_1092ncu_merged_pd.rename(columns = {'codeLength':'code_length'}, inplace = True)
viscode_1102ncu_merged_pd.rename(columns = {'codeLength':'code_length'}, inplace = True)
```

```python
viscode_merged_pd = viscode_1092ncu_merged_pd.append(viscode_1102ncu_merged_pd)
```
