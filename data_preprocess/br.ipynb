```python
import pandas as pd
import hashlib as hb
```

```python
br_1092ncu_a = pd.read_csv('./data/1092_NCU/1092_A-程式設計-Python_黃鈺晴教師.csv')
br_1092ncu_b = pd.read_csv('./data/1092_NCU/1092_B-程式設計-Python_黃鈺晴教師.csv')

br_1101ntut = pd.read_csv('./data/1101_NTUT/1101_程式設計-Python(北科)_黃鈺晴教師.csv')

br_1102ncu = pd.read_csv('./data/1102_NCU/1102_程式設計-Python_黃鈺晴教師.csv')
```

```python
br_1092ncu_a['class'] = 'a'
br_1092ncu_b['class'] = 'b'

br_1101ntut['class'] = 'c'
br_1102ncu['class'] = 'd'
```

```python
br_df = pd.merge(br_1092ncu_a, br_1092ncu_b, how = 'outer')
br_df = pd.merge(br_df, br_1101ntut, how = 'outer')
br_df = pd.merge(br_df, br_1102ncu, how = 'outer')
```

```python
br_df['username'] = br_df['username'].str.split('@',expand=True).get(0)
br_df = br_df[br_df['username'].str.isdigit() == True]
br_df = br_df[br_df['username'].apply(lambda x: len(str(x))>=9)]
```

```python
br_df = br_df.drop(columns=['Unnamed: 0', 'userid', 'roleid'])
```

```python
def md5(i):
    hl = hb.md5()
    
    hl.update(str(i).encode(encoding='utf-8'))

    #print('加密前為 ：' + str(i))
    #print('加密後為 ：' + hl.hexdigest())
    
    return hl.hexdigest()

br_df['username'] = br_df['username'].apply(md5)
```

```python
def is_contains_chinese(strs):
    for _char in strs:
        if '\u4e00' <= _char <= '\u9fa5':
            return True
    return False
```
