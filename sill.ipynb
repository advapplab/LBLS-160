```python
import pandas as pd
import numpy as np
import hashlib as hb
```

```python
sill_a = pd.read_csv('./data/問卷A班/程式語言學習策略量表A班.csv')
sill_b = pd.read_csv('./data/問卷B班/程式語言學習策略量表B班.csv')
```

```python
sill_a.rename(columns={'學號 (Student ID)': 'username'}, inplace=True)
sill_b.rename(columns={'學號 (Student ID)': 'username'}, inplace=True)
```

```python
sill_a['username'] = sill_a['username'].astype(str)
sill_b['username'] = sill_b['username'].astype(str)
```

```python
### MD5
def md5(i):
    hl = hb.md5()
    
    hl.update(str(i).encode(encoding='utf-8'))

    #print('加密前為 ：' + str(i))
    #print('加密後為 ：' + hl.hexdigest())
    
    return hl.hexdigest()
```

```python
sill_a['username'] = sill_a['username'].apply(md5)
sill_b['username'] = sill_b['username'].apply(md5)
```

```python
sill_df = pd.merge(sill_a, sill_b, how = 'outer')
```

```python
sill_df = sill_df.drop(columns=['時間戳記', '學校 (School)', '姓名(Name)'])
```

```python
sill_df = sill_df.fillna(0)
```

```python
def col(x):
    
    if x != 'username':
        x = x.split('.')[0]
        x = '#'+x

    return x
```

```python
# list(sill_df)[1].split('.')[0]

sill_df.rename(columns= lambda x: col(x),inplace = True)
```
